```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse
import time
import re

class AdvancedSchoolCrawler:
    def __init__(self, base_url, max_depth=1, delay=1):
        self.base_url = base_url.rstrip("/")
        self.domain = urlparse(base_url).netloc
        self.max_depth = max_depth
        self.delay = delay
        self.visited = set()
        self.queue = [(self.base_url, 0)]
        self.target_keywords = {"about", "admission", "curriculum", "results", "apply", "performance", "fees"}

    def fetch(self, url):
        try:
            headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
            res = requests.get(url, timeout=10, headers=headers)
            return res.text if res.status_code == 200 else None
        except:
            return None

    def clean_text(self, text):
        return " ".join(text.split())

    def extract_section_by_keywords(self, soup, keywords):
        """Heading ke niche ka data nikalne ke liye improved logic"""
        for heading in soup.find_all(['h1', 'h2', 'h3', 'h4', 'strong']):
            header_text = heading.get_text().lower()
            if any(k in header_text for k in keywords):
                content = []
                # Heading ke baad ke elements check karein
                for sibling in heading.find_all_next(['p', 'li', 'div'], limit=5):
                    txt = sibling.get_text(strip=True)
                    if len(txt) > 30 and not any(x in txt.lower() for x in ['cookie', 'javascript', 'meta']):
                        content.append(txt)
                if content:
                    return " ".join(content[:2])
        return "Refer to website"

    def extract_info(self, html):
        soup = BeautifulSoup(html, "html.parser")
        # Remove junk before processing
        for junk in soup(["script", "style", "noscript", "meta", "header", "footer"]):
            junk.decompose()
            
        full_text = soup.get_text(separator=" ", strip=True)

        # 1. About (Fixed 2 Paragraphs)
        paragraphs = [self.clean_text(p.get_text()) for p in soup.find_all('p') if len(p.get_text()) > 80]
        about_final = ". ".join(paragraphs[:2]) if paragraphs else "About info not found"

        # 2. Institution Details (Stricter Regex for short values)
        details = {}
        info_labels = ['Type', 'Ages', 'Students', 'Ratio', 'Founded', 'City', 'Annual fee']
        for label in info_labels:
            # Yeh regex ab label ke baad sirf 3-5 words uthayega, pure sentences nahi
            match = re.search(rf'{label}\s*[:\-]?\s*((?:\w+\s+){{0,4}}\w+)', full_text, re.IGNORECASE)
            details[label] = match.group(1).strip() if match else "N/A"

        # 3. Quotes (Filter out HTML noise)
        all_quotes = re.findall(r'["“]([^"”>]{40,250})["”]', full_text)
        final_quotes = [q.strip() for q in all_quotes if '<' not in q][:2]

        # 4. Specialized Sections
        return {
            "About": about_final,
            "Institution Details": details,
            "Quotes": final_quotes if final_quotes else ["No direct quotes found"],
            "How they teach": self.extract_section_by_keywords(soup, ['curriculum', 'teaching', 'learning', 'academic']),
            "Where students go": self.extract_section_by_keywords(soup, ['destination', 'outcome', 'results', 'alumni']),
            "Performance": self.extract_section_by_keywords(soup, ['performance', 'inspection', 'ofsted', 'reputation']),
            "Admission": self.extract_section_by_keywords(soup, ['process', 'how to apply', 'registration', 'entry'])
        }

    def crawl(self):
        while self.queue:
            url, depth = self.queue.pop(0)
            if url in self.visited or depth > self.max_depth: continue

            self.visited.add(url)
            print(f"\n{'='*50}\n🔍 SCRAPING: {url}")
            
            html = self.fetch(url)
            if html:
                data = self.extract_info(html)
                print(f"📖 ABOUT: {data['About']}\n")
                print(f"🏫 DETAILS: {data['Institution Details']}")
                print(f"💬 QUOTES: {data['Quotes']}")
                print(f"🎓 TEACHING: {data['How they teach']}")
                print(f"🚀 OUTCOMES: {data['Where students go']}")
                print(f"📊 PERFORMANCE: {data['Performance']}")
                print(f"📝 ADMISSION: {data['Admission']}")

                soup = BeautifulSoup(html, "html.parser")
                for a in soup.find_all("a", href=True):
                    link = urljoin(self.base_url, a["href"])
                    if self.domain in link and any(k in link.lower() for k in self.target_keywords):
                        if link not in self.visited:
                            self.queue.append((link, depth + 1))
            time.sleep(self.delay)

if __name__ == "__main__":
    crawler = AdvancedSchoolCrawler(base_url="http://wetherbyschool.co.uk", max_depth=1)
    crawler.crawl()
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse
import time
import re

class ProfessionalSchoolCrawler:
    def __init__(self, base_url, max_depth=1, delay=1):
        self.base_url = base_url.rstrip("/")
        self.domain = urlparse(base_url).netloc
        self.max_depth = max_depth
        self.delay = delay
        self.visited = set()
        self.queue = [(self.base_url, 0)]
        self.target_keywords = {"about", "admission", "curriculum", "results", "apply", "performance", "fees"}

    def fetch(self, url):
        try:
            headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
            res = requests.get(url, timeout=10, headers=headers)
            return res.text if res.status_code == 200 else None
        except:
            return None

    def clean_text(self, text, limit=350):
        cleaned = " ".join(text.split())
        if len(cleaned) > limit:
            return cleaned[:limit] + "..."
        return cleaned

    def extract_section_by_keywords(self, soup, keywords):
        """Behtar logic: Heading milne par uske niche ka limited text hi uthayega"""
        for heading in soup.find_all(['h1', 'h2', 'h3', 'h4', 'strong']):
            header_text = heading.get_text().lower()
            if any(k in header_text for k in keywords):
                content = []
                # Sirf agle 3 relevant paragraphs ya list items uthao
                for sibling in heading.find_all_next(['p', 'li'], limit=3):
                    txt = sibling.get_text(strip=True)
                    if len(txt) > 40:
                        content.append(txt)
                if content:
                    return self.clean_text(" ".join(content))
        return "Check specific page for details"

    def extract_info(self, html):
        soup = BeautifulSoup(html, "html.parser")
        
        # Clean Junk
        for junk in soup(["script", "style", "noscript", "nav", "footer", "header"]):
            junk.decompose()
            
        full_text = soup.get_text(separator=" ", strip=True)

        # 1. About Logic (Strict 2 Paras)
        paras = [p.get_text(strip=True) for p in soup.find_all('p') if len(p.get_text()) > 100]
        about_final = ". ".join(paras[:2]) if paras else "Main content not found"

        # 2. Institution Details (Stricter matching to avoid long sentences)
        details = {}
        info_labels = ['Type', 'Ages', 'Students', 'Ratio', 'Founded', 'City', 'Annual fee']
        for label in info_labels:
            # Regex optimized: Label ke baad maximum 4 words capture karega
            match = re.search(rf'{label}\s*[:\-]?\s*((?:\b\w+\b\s*){{1,4}})', full_text, re.IGNORECASE)
            details[label] = match.group(1).strip() if match else "N/A"

        # 3. Quotes logic: Filter out HTML-like strings
        quotes = re.findall(r'["“]([^"”]{40,200})["”]', full_text)
        final_quotes = [q.strip() for q in quotes if len(q.split()) > 5][:2]

        return {
            "About": self.clean_text(about_final, 500),
            "Institution Details": details,
            "Quotes": final_quotes if final_quotes else ["No direct quotes found"],
            "How they teach": self.extract_section_by_keywords(soup, ['curriculum', 'teaching', 'learning', 'academic', 'syllabus']),
            "Where students go": self.extract_section_by_keywords(soup, ['destination', 'outcome', 'university', 'exit', 'results']),
            "Performance": self.extract_section_by_keywords(soup, ['performance', 'inspection', 'rating', 'ofsted']),
            "Admission": self.extract_section_by_keywords(soup, ['process', 'how to apply', 'entry', 'registration'])
        }

    def crawl(self):
        while self.queue:
            url, depth = self.queue.pop(0)
            if url in self.visited or depth > self.max_depth: continue

            self.visited.add(url)
            print(f"\n{'='*60}\n🔍 TARGET: {url}\n{'='*60}")
            
            html = self.fetch(url)
            if html:
                data = self.extract_info(html)
                print(f"📖 ABOUT: {data['About']}\n")
                print(f"🏫 DETAILS: {data['Institution Details']}")
                print(f"💬 QUOTES: {data['Quotes']}")
                print(f"🎓 TEACHING: {data['How they teach']}")
                print(f"🚀 OUTCOMES: {data['Where students go']}")
                print(f"📊 PERFORMANCE: {data['Performance']}")
                print(f"📝 ADMISSION: {data['Admission']}")

                # Recursion for inner links
                soup = BeautifulSoup(html, "html.parser")
                for a in soup.find_all("a", href=True):
                    link = urljoin(self.base_url, a["href"])
                    if self.domain in link and any(k in link.lower() for k in self.target_keywords):
                        if link not in self.visited:
                            self.queue.append((link, depth + 1))
            time.sleep(self.delay)

if __name__ == "__main__":
    crawler = ProfessionalSchoolCrawler(base_url="http://wetherbyschool.co.uk", max_depth=1)
    crawler.crawl()
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse
import time
import re

class SmartSchoolCrawler:
    def __init__(self, base_url, delay=1):
        self.base_url = base_url.rstrip("/")
        self.domain = urlparse(base_url).netloc
        self.delay = delay
        self.visited = set()
        
        # Priority logic: In keywords waale links ko pehle visit karenge
        self.priority_keywords = ["about", "admission", "curriculum", "fees", "contact"]
        self.queue = [(self.base_url, 0)] 
        
        # Final Data Store
        self.extracted_data = {
            "About": None,
            "Established Year": None,
            "Admission Process": None,
            "Institution Details": None
        }

    def is_complete(self):
        """Check karega ki kya hume zaruri saari info mil gayi hai?"""
        # Agar About, Year aur Admission mil gaya hai toh true return karega
        return all([self.extracted_data["About"], 
                    self.extracted_data["Established Year"], 
                    self.extracted_data["Admission Process"]])

    def fetch(self, url):
        try:
            headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
            res = requests.get(url, timeout=10, headers=headers)
            return res.text if res.status_code == 200 else None
        except Exception as e:
            return None

    def extract_info(self, html, current_url):
        soup = BeautifulSoup(html, "html.parser")
        
        # Junk remove karna (Scripts, styles, nav)
        for junk in soup(["script", "style", "nav", "footer", "header"]):
            junk.decompose()

        full_text = soup.get_text(separator=" ", strip=True)

        # 1. About Section (Agar abhi tak nahi mila aur hum 'about' page par hain)
        if not self.extracted_data["About"] or "about" in current_url.lower():
            paras = [p.get_text(strip=True) for p in soup.find_all('p') if len(p.get_text()) > 100]
            if paras:
                self.extracted_data["About"] = " ".join(paras[:2])

        # 2. Established Year (Regex search)
        if not self.extracted_data["Established Year"]:
            year_match = re.search(r'(?:Founded|Established|Est\.|Since)\s*[:\-]?\s*(\d{4})', full_text, re.IGNORECASE)
            if year_match:
                self.extracted_data["Established Year"] = year_match.group(1)

        # 3. Admission Process (Special check for admission pages)
        if "admission" in current_url.lower() and not self.extracted_data["Admission Process"]:
            # Heading dhoondo jo Admission ya Apply se related ho
            target = soup.find(['h1', 'h2', 'h3'], text=re.compile(r'Admission|Apply|Process', re.I))
            if target:
                content = []
                for sib in target.find_all_next(['p', 'li'], limit=5):
                    content.append(sib.get_text(strip=True))
                self.extracted_data["Admission Process"] = " ".join(content)

    def crawl(self):
        print(f"🚀 Starting human-like search for: {self.base_url}")
        
        while self.queue and not self.is_complete():
            # URL nikalna
            url, depth = self.queue.pop(0)

            if url in self.visited or depth > 2:
                continue

            self.visited.add(url)
            print(f"🔍 Checking Page: {url}")

            html = self.fetch(url)
            if not html:
                continue

            # Information nikalna
            self.extract_info(html, url)

            # Agar sab mil gaya toh loop break kar do (Early Exit)
            if self.is_complete():
                print("✨ Required information found! Stopping search to save time.")
                break

            # Naye links dhoondna aur prioritize karna
            soup = BeautifulSoup(html, "html.parser")
            for a in soup.find_all("a", href=True):
                link = urljoin(self.base_url, a["href"])
                
                # Check internal link and not visited
                if self.domain in link and link not in self.visited:
                    # Agar link mein priority keywords hain toh queue ke aage daalo
                    if any(k in link.lower() for k in self.priority_keywords):
                        self.queue.insert(0, (link, depth + 1))
                    else:
                        self.queue.append((link, depth + 1))
            
            time.sleep(self.delay)

        # Final Results Print karna
        print("\n" + "="*50)
        print("🎯 FINAL EXTRACTED DATA")
        print("="*50)
        for key, value in self.extracted_data.items():
            print(f"🔹 {key}: {value if value else 'Not Found'}\n")

if __name__ == "__main__":
    # Test URL
    target_site = "http://wetherbyschool.co.uk"
    crawler = SmartSchoolCrawler(base_url=target_site)
    crawler.crawl()
```

```python
import asyncio
import csv
from playwright.async_api import async_playwright

async def run_scraper():
    async with async_playwright() as p:
        # Browser launch (headless=False kar sakte ho agar live dekhna hai)
        browser = await p.chromium.launch(headless=True)
        context = await browser.new_context(user_agent="Mozilla/5.0")
        page = await context.new_page()

        url = "https://www.wetherbyschool.co.uk/our-school/about-us"
        print(f"Navigating to: {url}...")

        try:
            # Page load wait
            await page.goto(url, wait_until="networkidle")
            
            # 1. Scraping Descriptive Paragraphs
            # Hum 'main' content area ke paragraphs target kar rahe hain
            paragraphs = await page.locator("main p").all_inner_texts()
            school_description = [p.strip() for p in paragraphs if len(p.strip()) > 60]

            # 2. Scraping Quotes
            quotes = await page.locator("blockquote").all_inner_texts()
            clean_quotes = [q.strip().replace('\n', ' ') for q in quotes]

            # 3. Scraping General Info (Title & Established Year hint)
            page_title = await page.title()
            # Logo ya header mein aksar established year hota hai
            header_text = await page.locator("header").inner_text()
            est_year = "75 Years of Excellence" if "75" in header_text else "Not found on page"

            # Results Print Karein
            print("\n" + "="*50)
            print(f"SCHOOL NAME: {page_title}")
            print(f"ESTABLISHED INFO: {est_year}")
            print("="*50)

            print("\n--- DESCRIPTION PARAGRAPHS ---")
            for i, para in enumerate(school_description[:5], 1): # Pehle 5 bade paragraphs
                print(f"{i}. {para}\n")

            # CSV mein Save karna
            with open('wetherby_school_data.csv', 'w', newline='', encoding='utf-8') as f:
                writer = csv.writer(f)
                writer.writerow(['Category', 'Content'])
                writer.writerow(['Title', page_title])
                writer.writerow(['Established', est_year])
                for p in school_description:
                    writer.writerow(['Description Para', p])
                for q in clean_quotes:
                    writer.writerow(['Quote', q])

            print(f"\nSuccess! Data saved to 'wetherby_school_data.csv'")

        except Exception as e:
            print(f"Error occurred: {e}")
        
        finally:
            await browser.close()

if __name__ == "__main__":
    asyncio.run(run_scraper())
```
